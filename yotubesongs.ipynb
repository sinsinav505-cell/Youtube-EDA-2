```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
yt_df = pd.read_csv('youtube-top-100-songs-2025.csv')
yt_df
```

```python
yt_df.head()
```

```python
#checking if there is any null values or not
n = yt_df.isnull().sum()
n
```

```python
#replacing the null values
yt_df.fillna('Unknown',inplace=True)
```

```python
n
```

```python

#frequency of different follower counts
sns.histplot(yt_df['channel_follower_count'])
```

```python
#sorting channel subscribers
yt_df.sort_values('channel_follower_count', ascending=False).head(10)
```

```python
#view count of each channel
plt.figure(figsize=(20,6))
sns.barplot(x = 'channel', y = 'view_count',data = yt_df)
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
yt_df
```

```python
#relation between channel view count and subscribers
sns.jointplot(x='view_count', y='channel_follower_count',data = yt_df,kind='reg')
```

```python
sns.jointplot(x='view_count', y='channel_follower_count',data = yt_df,kind='hex')
```

```python
#duration vs view count
sns.kdeplot(x='duration',y='view_count',data=yt_df,fill=True)
```

```python

sns.pairplot(yt_df, hue='categories',palette='Reds',diag_kind='kde')
```

```python
yt_df.head()
```

```python
sns.pairplot(yt_df, hue='categories',palette='Greens',diag_kind='hist')
```

```python
#which category have more view count?
plt.figure(figsize=(11,6))
sns.barplot(x='categories' , y='view_count' ,data= yt_df)
plt.tight_layout()
```

```python
#which category is more?
plt.figure(figsize=(11,6))
sns.countplot(x='categories',data=yt_df)
```
